```python
import requests
```

```python
url = "https://clinicaltrials.gov/api/v2/studies"
params = {
    "query.cond": "diabetes",
    "filter.overallStatus": "RECRUITING",
    "format": "json",
    "pageSize": 5
}

response = requests.get(url, params=params)
data = response.json()

# Mostrar títulos breves
for study in data.get("studies", []):
    title = study.get("protocolSection", {}).get("identificationModule", {}).get("briefTitle")
    print(title)
```

```python
# Por condición específica y pais 

url = "https://clinicaltrials.gov/api/v2/studies"
params = {
    "query.cond": "diabetes",
    "query.locn": "Argentina",
    "filter.overallStatus": "RECRUITING",
    "fields": "NCTId,BriefTitle,OverallStatus",
    "format": "json",
    "pageSize": 5
}

response = requests.get(url, params=params)
data = response.json()

for study in data.get("studies", []):
    print(study.get("protocolSection", {}).get("identificationModule", {}).get("briefTitle"))

```
